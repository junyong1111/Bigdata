# 1번 문제. ANOVA(분삭분석)
다음은 A, B, C 그룹 인원 성적 데이터이다.
세 그룹의 성적 평균이 같다고 할 수 있는지 ANOVA 분석을 실시하시오.
before : 혈압약을 먹기 전
after :  혈압약을 먹은 후
- 귀무가설(H0) : A,B,C 그룹의 평균이 같다.
- 대립가설(H1) : A,B,C 그룹의 평균이 다르다.
- 유의수준(Alpha) = 0.05

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
from scipy.stats import shapiro
```

```python
df = pd.DataFrame( {
    'A': [120, 135, 122, 124, 135, 122, 145, 160, 155, 142, 144, 135, 167],
    'B' : [110, 132, 123, 119, 123, 115, 140, 162, 142, 138, 135, 142, 160],
    'C' : [130, 120, 115, 122, 133, 144, 122, 120, 110, 134, 125, 122, 122]})
print(df.head(3))
```

```python
H0 = " A,B,C 그룹의 평균이 같다."
H1 = "A,B,C 그룹의 평균이 다르다"
alpha = 0.05
```

```python
#-- 1. 정규성 검정(쌍체의 경유 두 집단의 차이에 대한 정규성)
a_s, a_p = shapiro(df['A'])
b_s, b_p = shapiro(df['B'])
c_s, c_p = shapiro(df['C'])

if a_p > alpha and b_p > alpha and c_p > alpha:
  print('정규성 채택')
else:
  print('정규성 기각')
```

```python
#-- 2. 등분산성 검정
s,p = stats.bartlett(df['A'], df['B'], df['C'])

if p > alpha :
  print("등분산성을 따른다")
else:
  print("등분산성을 따르지 않는다.")
```

```python
#-- 3. 통계량 계산
statics, p_value = stats.f_oneway(df['A'], df['B'], df['C'])
print(statics.round(4), p_value.round(4))
#-- 3-1. 정규성을 따르지 않는 경우
c_s, c_v = stats.kruskal(df['A'], df['B'], df['C'])
print(c_s.round(4), c_v.round(4))

if p_value > alpha:
  print("귀무가설 채택")
  print(H0)
else:
  print("귀무가설 기각")
  print(H1)
```

# 1-2번 문제. 1-2 데이터 형태가 다를 경우


```python
df2 = pd.DataFrame( {
'항목': [
    'A','A','A','A','A','A','A','A','A','A','A','A','A',
    'B','B','B','B','B','B','B','B','B','B','B','B','B',
    'C','C','C','C','C','C','C','C','C','C','C','C','C',],
    'value': [120, 135, 122, 124, 135, 122, 145, 160, 155, 142, 144, 135, 167,
             110, 132, 123, 119, 123, 115, 140, 162, 142, 138, 135, 142, 160,
             130, 120, 115, 122, 133, 144, 122, 120, 110, 134, 125, 122, 122]
  })
print(df2.head(3))
```

```python
# 각각 필터링해서 변수명에 저장하고 분석 진행
a = df2[ df2['항목']=='A' ]['value']
b = df2[ df2['항목']=='B' ]['value']
c = df2[ df2['항목']=='C' ]['value']

# 분산분석(F_oneway)
import scipy.stats as stats
statistic, pvalue = stats.f_oneway(a, b, c)
print(round(statistic,4), round(pvalue,4) )
```
